```python
import numpy as np
import matplotlib.pyplot as plt
```

For simplicity, we'll generate synthetic data.

```python
np.random.seed(42)  # For reproducibility
X = np.random.rand(100, 1)  # 100 samples, single feature
y = 5 * X.squeeze() + 2 + np.random.randn(100) * 0.5  # Linear relationship with noise

# Visualize the data
plt.scatter(X, y, color='blue', alpha=0.6, label='Data points')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Synthetic Data')
plt.legend()
plt.show()
```

Add Bias Term to X

We need to include an intercept term in the regression model by augmenting 
𝑋
X with a column of ones.

```python
# Add bias term (column of ones)
X_aug = np.hstack([np.ones((X.shape[0], 1)), X])
X_aug
```

We solve 𝛽=(𝑋⊤𝑋)−1𝑋⊤𝑦β=(X ⊤ X) −1 X ⊤ y.

```python
# Calculate beta using the normal equation
X_transpose = X_aug.T
beta = np.linalg.inv(X_transpose @ X_aug) @ X_transpose @ y

beta  # beta[0] = intercept, beta[1] = slope
```

Compute predictions for the fitted model.

```python
# Predictions
y_pred = X_aug @ beta
y_pred
```

Plot the original data and the fitted line.

```python
plt.scatter(X, y, color='blue', alpha=0.6, label='Data points')
plt.plot(X, y_pred, color='red', label='Fitted line', linewidth=2)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Univariate Linear Regression Using Vector Algebra')
plt.legend()
plt.show()
```
